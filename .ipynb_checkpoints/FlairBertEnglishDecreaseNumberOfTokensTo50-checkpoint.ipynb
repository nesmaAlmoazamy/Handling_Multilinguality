```python
from flair.data_fetcher import NLPTaskDataFetcher
from flair.embeddings import WordEmbeddings, FlairEmbeddings, DocumentLSTMEmbeddings,DocumentRNNEmbeddings, DocumentPoolEmbeddings
from flair.models import TextClassifier
from flair.trainers import ModelTrainer
from pathlib import Path
corpus = NLPTaskDataFetcher.load_classification_corpus(Path('./'), test_file='test.csv', dev_file='dev.csv', train_file='train.csv')
```

```python
len(corpus.train)
```

```python
from flair.embeddings import BertEmbeddings

word_embeddings = [BertEmbeddings('bert-base-cased')]


```

```python
document_embeddings: DocumentRNNEmbeddings = DocumentLSTMEmbeddings(word_embeddings, hidden_size=512, reproject_words=True, reproject_words_dimension=256)
document_embeddings

```

```python
# 5. create the text classifier
classifier = TextClassifier(document_embeddings, label_dictionary=corpus.make_label_dictionary(), multi_label=False)

classifier
```

```python
# # 6. initialize the text classifier trainer
trainer = ModelTrainer(classifier, corpus)

trainer
```

```python
# # 7. start the training
trainer.train('./',
              learning_rate=0.1,
              patience=2,
              max_epochs=10)
```
