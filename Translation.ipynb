```python
import seaborn as sn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
def removes_url(text):
    text = text.apply(lambda x: re.split('https:\/\/.*', str(x))[0])
    return text

def remove_url(text):
    text = text.apply(lambda x: re.split('http:\/\/.*', str(x))[0])
    return text

dataset['text'] = removes_url(dataset['text'])
dataset['text'] = remove_url(dataset['text'])

dataset.head()
```

```python
def remove_numbers(text):
    text = text.str.replace('\d+', '')
    return text


dataset['text'] = remove_numbers(dataset['text'])
dataset.head()
```

```python
def remove_punctuations(text):
    text = text.str.replace('[^\w\s]','')
    return text


dataset['text'] = remove_punctuations(dataset['text'])
dataset.head()
```

```python
def remove_blank_space(col):
    col = col.str.strip()
    col = col.replace('\s+', ' ', regex=True)   
    return col


dataset['text'] = remove_blank_space(dataset.text)
dataset.head()
```
